```python
!pip install openai
import openai

class ScientificResearchAssistant:
    def __init__(self, api_key, model_name="gpt-3.5-turbo", max_tokens=1000, temperature=0.2):
        openai.api_key = api_key
        self.model_name = model_name
        self.max_tokens = max_tokens
        self.temperature = temperature

    def _generate_chat_completion(self, prompt):
        messages = [
            {"role": "system", "content": "I am an intelligent assistant capable of proposing scientific hypotheses and understanding experimental plans. Please respond in Korean."},
            {"role": "assistant", "content": f"What website would you recommend for learning more about this hypothesis and research plan in greater detail, preferably one with explanations included?"},
            {"role": "user", "content": f"The following prompt requires a master's or doctoral level of knowledge as it will be used for research in scientific concepts.: '{prompt}'"}
        ]
        completion = openai.ChatCompletion.create(
            model=self.model_name,
            messages=messages,
            max_tokens=self.max_tokens,
            temperature=self.temperature
        )
        return completion.choices[0].message["content"].strip()

    def suggest_hypothesis(self, initial_idea):
        prompt_text = f"Based on the initial idea '{initial_idea}', please suggest a scientific hypothesis in Korean."
        return self._generate_chat_completion(prompt_text)

    def plan_experiment(self, hypothesis):
        prompt_text = f"Please propose an experimental plan to verify the hypothesis '{hypothesis}' in Korean."
        return self._generate_chat_completion(prompt_text)

def chatbot_interface():
    assistant = ScientificResearchAssistant(api_key='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')
    print("안녕하세요. 저는 당신의 학술 연구를 돕는 AI어시스턴트, 레비(laby)입니다. 저는 양질의 데이터로 메타분석을 할 수 있어요.")
    print("무엇을 도와드릴까요?")
    while True:
        print("옵션을 선택해주세요:")
        print("가설 정립")
        print("연구 계획 수립")
        choice = input("선택하세요 (가설 정립/연구 계획 수립): ")

        if choice == "가설 정립":
            initial_idea = input("초기 아이디어를 입력하세요: ")
            hypothesis = assistant.suggest_hypothesis(initial_idea)
            print("제안된 가설:", hypothesis)

        elif choice == "연구 계획 수립":
            hypothesis_input = input("가설을 입력하세요: ")
            experimental_plan = assistant.plan_experiment(hypothesis_input)
            print("실험 계획:", experimental_plan)
            break

        else:
            print("잘못된 선택입니다. 가설 정립, 연구 계획 수립을 선택해주세요.")

if __name__ == "__main__":
    chatbot_interface()
```
